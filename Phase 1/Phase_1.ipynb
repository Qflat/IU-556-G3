# Abstract



This project will create models that aim to predict who will default on their home credit loans. The data being used as our test datasets comes from the Home Credit Default Risk (HCDR) online competition. The main dataset is the application test data, which has data on whether the loan was repaid or not and information on the client. Other datasets that will be used include bureau, bureau_balance, previous_application, POS_CASH_BALANCE, credit_card_balance, and installments_payment. Using this data, we plan to develop machine learning pipelines that encompass data cleaning, feature encoding, and feature scaling. These pipelines will then be used to perform data preprocessing, feature engineering, and data modeling. As a result of this, our goal of this project is to accurately predict whether a client applying for a loan will repay it. Should this goal prove successful, the deployment of this model could be employed by various financial institutions in pursuit of better business and loaning practices. 



# Data Description


We plan to use the data sets that are provided by the HCDR Kaggle Competition file. These include:
Application Data: Includes the majority of information regarding clients, such as the gender, income, family status, education, number of children, amount of income, and whether the client owns a car or not. The application data has split the main training data from the testing data.
Bureau Data: This dataset contains data on clients’ credit history.
Bureau_Balance Data: This dataset breaks down the credit history by months.
Previous_Application Data: Contains information on clients who have had previous applications for loans from Home Credit.
POS_CASH_BALANCE Data: Breaks down point of sale and cash loan data by months.
Credit_Card_Balance Data: Contains data for clients who have had a Home Credit credit card.
Installments_Payment Data: Includes payment histories for clients who have had a previous Home Credit loan.


# Machine Algorithms and Metrics:


For the Home Credit Default Risk (HCDR) project, our goal is to predict whether or not a loan applicant will repay their loan using demographic, financial, and behavioral data. Because these datasets have many inconsistencies, we plan on using efficient machine learning algorithms that are designed for tabular classification. 

The machine learning algorithms we will focus on for this project are logistic regression, random forest, and three different forms of gradient boosting, which are Light Gradient Boosting Machine (LGBM), Extreme Gradient Boosting (XGBoost), and Categorical Boosting (CatBoost). Logistic regression will be used to analyze performance. Random Forest will be used to understand non-linear relationships and feature characteristics. The gradient boosting frameworks will be used to improve accuracy and understand complex features. 

To measure the performance of these algorithms, our team will mainly focus on using AUC-ROC, which is the official Kaggle competition metric. Our team will also use other metrics that have been discussed thoroughly throughout the class, such as Precision, Recall, and F1-score. These metrics will be used to evaluate the relationships between false positives and false negatives in the datasets. We will also use Precision-Recall AUC and the Gini Coefficient to assess the model’s ability to rank applicants by credit risk and measure its overall discriminatory power in predicting defaults. 

Through using these algorithms and metrics, our team will build a comprehensive modeling framework capable of effectively handling imbalanced credit data, optimizing predictive accuracy, and providing actionable insights for fair lending decisions.



# ML Pipelines


Logistic Regression: Predicts a categorical dependent variable based on one or more independent variables
Random Forest: Uses an ensemble method to create multiple decision trees for improved accuracy
LGBM: A fast, high-performance gradient boosting framework that uses tree-based learning algorithms optimized for efficiency and scalability
XGBoost: Uses a sequence of decision trees (similar to random forest) to build a predictive model
CatBoost: By using a random ordering of the dataset, a categorical feature is added to each permutation and assigned a numerical value. The algorithm makes sure to not include future values in an effort to prevent data leakage
